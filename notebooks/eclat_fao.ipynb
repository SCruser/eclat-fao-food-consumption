```python
import pandas as pd
from google.colab import files

uploaded = files.upload()  # selectezi consumption_user.csv
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)
print("Shape (linii, coloane):", df.shape)

display(df.head(10))
print("\nColoane:", list(df.columns))

print("\nInfo:")
df.info()

print("\nValori lipsă pe coloană:")
print(df.isna().sum().sort_values(ascending=False).head(20))
```

```python
# Pasul 2 – Selectare coloane relevante
cols = ["SUBJECT", "SURVEY_DAY", "INGREDIENT_ENG"]
df_clean = df[cols].copy()

# Eliminăm rândurile fără ingredient
df_clean = df_clean.dropna(subset=["INGREDIENT_ENG"])

# Normalizăm textul ingredientelor
df_clean["INGREDIENT_ENG"] = (
    df_clean["INGREDIENT_ENG"]
    .str.lower()
    .str.strip()
)

print("Shape după curățare:", df_clean.shape)
display(df_clean.head(10))
```

```python
print("Număr ingrediente unice:", df_clean["INGREDIENT_ENG"].nunique())
print("Top 10 ingrediente cele mai frecvente:")
print(df_clean["INGREDIENT_ENG"].value_counts().head(10))
```

```python
# Pasul 3 – Construire tranzacții

transactions = (
    df_clean
    .groupby(["SUBJECT", "SURVEY_DAY"])["INGREDIENT_ENG"]
    .apply(lambda x: sorted(set(x)))  # set = eliminăm duplicatele
)

print("Număr tranzacții:", len(transactions))
print("\nExemplu de tranzacție:")
transactions.iloc[0]
```

```python
# Lungimea tranzacțiilor
transaction_lengths = transactions.apply(len)

print("Statistici tranzacții:")
print(transaction_lengths.describe())
```

```python
# Pasul 4.1 – Construim reprezentarea verticală (ECLAT)

from collections import defaultdict

# Indexăm tranzacțiile
transaction_list = list(transactions)
transaction_ids = list(range(len(transaction_list)))

# ingredient -> set de transaction IDs
vertical_db = defaultdict(set)

for tid, items in zip(transaction_ids, transaction_list):
    for item in items:
        vertical_db[item].add(tid)

print("Număr ingrediente în vertical DB:", len(vertical_db))

# Exemplu
example_item = list(vertical_db.keys())[0]
print("Exemplu ingredient:", example_item)
print("Apare în", len(vertical_db[example_item]), "tranzacții")
```

```python
min_support = 0.05
min_support_count = int(min_support * len(transaction_list))
min_support_count
```

```python
# Pasul 4.3 – ECLAT (itemsets frecvente de dimensiune 1 și 2)

frequent_itemsets = []

# Itemsets de dimensiune 1
freq_items_1 = {
    frozenset([item]): tids
    for item, tids in vertical_db.items()
    if len(tids) >= min_support_count
}

print("Itemsets frecvente (1 item):", len(freq_items_1))

frequent_itemsets.extend([
    (itemset, len(tids)/len(transaction_list))
    for itemset, tids in freq_items_1.items()
])

# Itemsets de dimensiune 2
items = list(freq_items_1.items())

for i in range(len(items)):
    itemset_i, tids_i = items[i]
    for j in range(i+1, len(items)):
        itemset_j, tids_j = items[j]
        intersection = tids_i & tids_j
        if len(intersection) >= min_support_count:
            frequent_itemsets.append((
                itemset_i | itemset_j,
                len(intersection)/len(transaction_list)
            ))

print("Total itemsets frecvente (1 și 2):", len(frequent_itemsets))
```

```python
# Sortăm după support
frequent_itemsets_sorted = sorted(
    frequent_itemsets,
    key=lambda x: x[1],
    reverse=True
)

# Primele 15
for itemset, supp in frequent_itemsets_sorted[:15]:
    print(f"{set(itemset)} -> support = {supp:.3f}")
```

```python
# Dicționar: itemset -> support
support_dict = {
    itemset: supp
    for itemset, supp in frequent_itemsets
}
```

```python
rules = []

min_confidence = 0.6

for itemset, supp in frequent_itemsets:
    if len(itemset) == 2:
        A, B = list(itemset)

        # A -> B
        conf_ab = supp / support_dict[frozenset([A])]
        lift_ab = conf_ab / support_dict[frozenset([B])]

        if conf_ab >= min_confidence:
            rules.append((A, B, supp, conf_ab, lift_ab))

        # B -> A
        conf_ba = supp / support_dict[frozenset([B])]
        lift_ba = conf_ba / support_dict[frozenset([A])]

        if conf_ba >= min_confidence:
            rules.append((B, A, supp, conf_ba, lift_ba))

print("Număr reguli generate:", len(rules))
```

```python
# Sortăm după lift
rules_sorted = sorted(rules, key=lambda x: x[4], reverse=True)

# Top 15 reguli
for A, B, supp, conf, lift in rules_sorted[:15]:
    print(f"{A} -> {B} | support={supp:.3f}, confidence={conf:.3f}, lift={lift:.3f}")
```
